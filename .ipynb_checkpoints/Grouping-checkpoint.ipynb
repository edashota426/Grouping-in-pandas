```python
#inport pandas & numpy
import pandas as pd
import numpy as np

#import csv and  convert it to pandas dateframe
df = pd.read_csv("sample.csv", encoding="utf_8")
#add calum for "group"
df["group"] = int(100)
#sort the dateframe by "skill"
sorted_df = df.sort_values(by='skill')
#devide dateframe into "male" & "female"
male = sorted_df[sorted_df.gender == "M"]
female = sorted_df[sorted_df.gender == "F"]
```

```python
#set group number
group_number = 5
#
list(range(group_number))
male["group"]  = list(range(group_number))*( int(len(male)/group_number)) + list(range(int(len(male)%group_number)))
female["group"]  = list(range(group_number))[::-1]*( int(len(female)/group_number)) + list(range(group_number))[::-1][:len(female)%group_number]

labeled_df = pd.concat([male, female])
sorted_labeled_df = labeled_df = labeled_df.sort_values(by="group")
sorted_labeled_df
```
